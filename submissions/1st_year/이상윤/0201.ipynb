# 📊 데이터 시각화 기초: Matplotlib

데이터 분석에 있어서 중요한 능력 중 하나는 **'데이터를 시각화하는 것'**입니다. 
숫자로만 된 데이터를 그래프로 그리면, 데이터의 특징이나 규칙(패턴)을 한눈에 파악할 수 있습니다.
파이썬에서 가장 유명하고 강력한 시각화 도구인 `matplotlib`을 사용해 봅시다.

```python
# matplotlib 라이브러리 안에서 그래프를 그리는 모듈인 pyplot을 불러옵니다.
# 이름이 너무 기니까 'plt'라는 별명으로 줄여서 부르겠습니다.
import matplotlib.pyplot as plt
```

---
## 1. 선 그래프 (Line Plot) 그리기
가장 기본이 되는 선 그래프입니다. 시간에 따른 변화(예: 주식 가격, 온도 변화 등)를 보여줄 때 많이 사용합니다.

- `plt.plot(x, y)`: x축과 y축 데이터를 넣어 선 그래프를 그립니다.
- `plt.show()`: 그려진 그래프를 화면에 보여줍니다.

```python
# x축 데이터 (예: 1일, 2일, 3일, 4일)
days = [1, 2, 3, 4]

# y축 데이터 (예: 공부 시간)
study_hours = [2, 4, 3, 6]

# 그래프 그리기
plt.plot(days, study_hours)

# 화면에 출력
plt.show()
```

---
## 2. 그래프 예쁘게 꾸미기
그래프만 덩그러니 있으면 어떤 데이터인지 알 수 없습니다. 제목(Title), 축 이름(Label), 그리고 범례(Legend)를 추가해 봅시다.
*(주의: 기초 단계에서는 한글 깨짐을 방지하기 위해 영어로 작성합니다.)*

```python
days = [1, 2, 3, 4]
study_hours = [5, 3, 4, 0]
play_hours = [5, 7, 6, 10]

# 두 개의 선을 하나의 그래프에 그릴 수 있습니다. label을 달아줍니다.
plt.plot(days, study_hours, label="game", color="blue")
plt.plot(days, play_hours, label="study", color="red")

# 제목과 축 이름 추가
plt.title("exam period")
plt.xlabel("Days")
plt.ylabel("Hours")

# 범례(Legend) 표시
plt.legend()

# 그래프 보여주기
plt.show()
```

---
## 3. 점으로 데이터 흩뿌리기: 산점도 (Scatter Plot)
머신러닝과 AI를 할 때 가장 많이 보는 그래프입니다. 데이터들이 어떻게 모여 있는지(군집), 어떤 상관관계가 있는지 점을 찍어 확인합니다.

- `plt.scatter(x, y)`를 사용합니다.

```python
# 키(Height)와 몸무게(Weight) 데이터라고 가정해 봅시다.
heights = [150, 160, 165, 170, 175, 180, 185]
weights = [50, 55, 60, 65, 75, 70, 85]

# 산점도 그리기
plt.scatter(heights, weights, color="green")

plt.title("Height vs Weight")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")

plt.show()
```

3. `plt.bar(x, y)`를 사용하여 막대 그래프를 그리세요.
4. 제목은 "My Fruits", x축은 "Fruits", y축은 "Preference"로 설정하세요.

```python
# 여기에 코드를 작성하세요.
fruits = ["apple", "banana", "grape", "orange"]
preference = ["8", "6", "7", "9"]
plt.plot(fruits, preference)

plt.title("my_fruits_preference")
plt.xlabel("fruits")
plt.ylabel("preference")

plt.legend()

plt.show()
```
